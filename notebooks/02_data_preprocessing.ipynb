```python
import pandas as pd

from config.paths import STAGING_DIR
from src.etl.match_info import process_match_info
```

```python
match_df = pd.read_csv(
    STAGING_DIR / "dataset_metadata.csv"
)

print(match_df.shape)

match_df.head()
```

```python
match_df.info()

match_df.describe(include="all").T

match_df.isna().sum()
```

```python
processed_match_df = process_match_info(
    match_df
)
```

```python
processed_match_df.head()

processed_match_df.info()

processed_match_df.columns
```

```python
print("Raw Shape")
print(match_df.shape)

print()

print("Processed Shape")
print(processed_match_df.shape)

new_columns = (
    set(processed_match_df.columns)
    - set(match_df.columns)
)

print(sorted(new_columns))
```

```python
output_path = (
    STAGING_DIR
    / "match_info_processed.parquet"
)

processed_match_df.to_parquet(
    output_path,
    index=False,
)

print(output_path)
```
