# Notebook 06 – Analytics Framework

## Objective

This notebook transforms the relational IPL warehouse into reusable business analytics.

Unlike Notebook 03 (Exploratory Analysis), this notebook operates entirely on the MySQL Star Schema constructed in Notebook 05.

The notebook serves as a thin orchestration layer that delegates all business logic to reusable modules within the `src` package.

---

## Workflow

1. Validate environment
2. Load project configuration
3. Execute analytics pipeline
4. Display executive summary
5. Display domain-specific analytics
6. Export analytics artifacts
7. Display execution summary

```python
from pathlib import Path

import pandas as pd

from config.config import (
    PROJECT_ROOT,
    ANALYTICS_DATA_DIR,
)

from src.workflows.analytics_pipeline import (
    run_analytics_pipeline,
)

from src.exports.analytics_exporter import (
    export_analytics_results,
)
```

```python
print("=" * 80)
print("IPL Analytics & Strategy Platform")
print("Notebook 06 - Analytics Framework")
print("=" * 80)

print(f"Project Root : {PROJECT_ROOT}")
print(f"Analytics Dir: {ANALYTICS_DATA_DIR}")

assert Path(PROJECT_ROOT).exists(), "Project root not found."
assert Path(ANALYTICS_DATA_DIR).exists(), "Analytics directory not found."

print("\nEnvironment validation completed successfully.")

EXPORT_RESULTS = True

DISPLAY_ANALYTICS = True
```

```python
pipeline_results = run_analytics_pipeline()

print()

print("=" * 80)
print("Analytics Pipeline Completed")
print("=" * 80)

print(f"Status         : {pipeline_results.status}")
print(f"Execution Time : {pipeline_results.execution_time:.2f} seconds")
```

```python
print("=" * 80)
print("EXECUTIVE SUMMARY")
print("=" * 80)

executive_summary = pipeline_results.executive_summary

summary_df = (
    pd.DataFrame(
        executive_summary.items(),
        columns=["Metric", "Value"],
    )
)

display(summary_df)
```

```python
print("=" * 80)
print("SEASON ANALYTICS")
print("=" * 80)

season_results = pipeline_results.analytics_results["season"]

for name, dataframe in season_results.items():

    print()

    print(name.replace("_", " ").title())

    display(dataframe)
```

```python
print("=" * 80)
print("TEAM ANALYTICS")
print("=" * 80)

team_results = pipeline_results.analytics_results["team"]

for name, dataframe in team_results.items():

    print()

    print(name.replace("_", " ").title())

    display(dataframe)
```

```python
print("=" * 80)
print("VENUE ANALYTICS")
print("=" * 80)

venue_results = pipeline_results.analytics_results["venue"]

for name, dataframe in venue_results.items():

    print()

    print(name.replace("_", " ").title())

    display(dataframe)
```

```python
print("=" * 80)
print("PLAYER ANALYTICS")
print("=" * 80)

player_results = pipeline_results.analytics_results["player"]

for name, dataframe in player_results.items():

    print()

    print(name.replace("_", " ").title())

    display(dataframe)
```
