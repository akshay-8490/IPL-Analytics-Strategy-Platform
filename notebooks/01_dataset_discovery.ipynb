### Objectives

- Validate raw IPL dataset
- Discover available match files
- Extract metadata from Cricsheet info files
- Perform data quality validation
- Generate discovery artifacts

```python
from config.config import *

from src.discovery import (
    build_dataset_catalog,
    build_dataset_metadata,
    build_discovery_report
)

from src.validation.quality_checks import (
    validate_dataset
)

from src.exports import (
    export_discovery_artifacts
)

from src.utils.logging_utils import (
    configure_logger,
    get_logger
)
```

```python
configure_logger()

logger = get_logger(__name__)

logger.info("Notebook 01 started.")
```

```python
catalog_df = build_dataset_catalog(
    RAW_DATA_DIR
)

catalog_df.head()
```

```python
metadata_df = build_dataset_metadata(
    catalog_df
)

metadata_df.head()
```

```python
validation_report = validate_dataset(
    catalog_df,
    metadata_df
)

validation_report
```

```python
report_df = build_discovery_report(
    validation_report
)

report_df
```

```python
export_discovery_artifacts(
    catalog_df=catalog_df,
    metadata_df=metadata_df,
    report_df=report_df,
    output_directory=STAGING_DIR
)
```

```python
print("=" * 60)

print("IPL DATASET DISCOVERY SUMMARY")

print("=" * 60)

for metric, value in validation_report.items():

    print(f"{metric:<30} : {value}")

print("=" * 60)

logger.info("Notebook 01 completed successfully.")
```
